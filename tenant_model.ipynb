```python
def clean_income(value):
    if 'k' in value:
        value = value.replace('k', '000')
        
    if 'K' in value:
        value = value.replace('K', '000')
        
    if 'thasand' in value:
        value = value.replace('thasand', '000')

    if "\'" in value:
        value = value.replace(value, '28000')

#typo in entry
    if  value == '560000':
        value = value.replace(value, '56000')
        
#     if value == '27000-30000':
#         value = value.replace(value, '28500')
        
#     if value == '30000-40000':
#         value = value.replace(value, '35000')
        
#     if value == '25000-30000':
#         value = value.replace(value, '27500')
        
    if '-' in value:
        split_parts = value.split('-')
        value = value.replace(value, split_parts[1])
#         value = value.replace(value, '27500')

    return value

# # Apply the cleaning function to the 'individual_income' column
tenant['individual_income'] = tenant['individual_income'].apply(clean_income)
```
